## 1.1. Building dataframes from scratch

```python
data = {
    'weekdays': ['Sun', 'Sun', 'Mon', 'Mon'],
    'cities': ['Austin', 'Dallas', 'Austin', 'Dallas'],
    'visitors': [139, 237, 326, 456],
    'signups': [7, 12, 3, 5]
}

users = pd.DataFrame(data)
users
```

```python
weekdays = ['Sun', 'Sun', 'Mon', 'Mon']
cities = ['Austin', 'Dallas', 'Austin', 'Dallas']
visitors = [139, 237, 326, 456]
signups = [7, 12, 3, 5]

list_labels = ['weekdays','cities','visitors','signups']
list_cols = [weekdays,cities,visitors,signups]
zipped = dict(list(zip(list_labels, list_cols)))
zipped
```

```python
users = pd.DataFrame(zipped)
users
```

```python
users['fees'] = 0
users
```
